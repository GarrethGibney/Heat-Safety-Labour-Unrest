```python
# Select only CA
states_to_plot = ["CA"]

# Compute average of all other states
other_states = counts.drop(columns=states_to_plot, errors="ignore")
counts["Other States Avg"] = other_states.mean(axis=1)

# Restrict period to 2000–2010
subset_period = counts.loc[(counts.index >= 2000) & (counts.index <= 2010),
                           states_to_plot + ["Other States Avg"]]

# Plot CA vs Other States Avg
subset_period.plot(figsize=(12,6))

plt.title("Frequency of Observations by Year (2000–2010: CA vs Other States Avg)")
plt.xlabel("Year")
plt.ylabel("Count of Observations")

# Add vertical reference lines
plt.axvline(x=2005, color="black", linestyle="--", linewidth=1.2,
            label="2005 California Outdoor Heat Safety Law")
plt.axvline(x=2006, color="black", linestyle="-", linewidth=1.2,
            label="2006 California Outdoor Heat Safety Law (Effective)")

# Show legend
plt.tight_layout()
plt.show()
```

```python
# Define hot states
hot_states = ["AZ", "NV", "TX", "NM", "FL"]
states_to_plot = ["CA"] + hot_states

# Subset counts
subset = counts[states_to_plot].copy()

# Compute the average of hot states (excluding CA)
subset["Hot States Avg"] = subset[hot_states].mean(axis=1)

# Restrict period to 2000–2010
subset_period = subset.loc[(subset.index >= 2000) & (subset.index <= 2010)]

# Plot CA vs Hot States Avg
subset_period[["CA", "Hot States Avg"]].plot(figsize=(12,6), linewidth=2)

plt.title("Frequency of Observations by Year (2000–2010: CA vs Hot States Avg)")
plt.xlabel("Year")
plt.ylabel("Count of Observations")

# Add vertical reference lines
plt.axvline(x=2005, color="black", linestyle="--", linewidth=1.2,
            label="2005 California Outdoor Heat Safety Law (Vote)")
plt.axvline(x=2006, color="black", linestyle="-", linewidth=1.2,
            label="2006 California Outdoor Heat Safety Law (Effective)")

# Show legend
plt.legend(title="State / Reference Lines", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
